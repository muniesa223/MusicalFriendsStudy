```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
#df = pd.read_csv('similarUsers.csv')
#data= data.str.split(",", n = 1, expand = True) 
#df.head()
#df = pd.read_csv('similarUsers.csv')
x = []
y = []



with open('similarUsers.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(str(row[0]))
        y.append(str(row[1]))
        

        

fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]

fig_size[0] = 30
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size




my_plot=plt.bar(x,y, label='Simmilarity', align='center')
plt.xticks(rotation='vertical')
plt.xlabel('USERS')
plt.ylabel('JACCARD SIMILARIRTY')
plt.title('Similarity')
plt.legend()
plt.show()
plt.savefig('similarFriends.png')


#data=str(df['name'])
#print(data)

#my_plot = df.plot(kind='bar',title="Most Similar Users",tick_label=data[8])
#my_plot.set_xlabel("Users")
#my_plot.set_ylabel("JaccSim")
```
